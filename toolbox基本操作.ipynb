```python
# 
# 
import toolbox
import cv2

# pngalpha
tim = toolbox.imread("testdata/icon.png", cv2.IMREAD_UNCHANGED)
print("width, height:", tim.size)
print("shape:", tim.shape)

# alpha
alpha = cv2.split(tim.data)[3]
toolbox.show_image([tim, alpha]) #[tim, alpha])
```

```python
# np.zeros 
# np.ones 
import numpy as np
import toolbox
black = np.zeros((10, 10, 3), np.uint8)
white = np.ones((10, 10, 3), np.uint8) * 255
toolbox.show_image([black, white], ['black', 'white'])
```
